# What's the Big Deal with Data Visualization?

There are two different reasons that you may use data visualization:
* Exploratory Data Analysis (EDA) and data cleaning - trying to understand the data for yourself
* Communicating results of your analytics work to teammates or business stakeholders

When you are creating charts for yourself, you can do whatever you want. They don't have to look pretty, and you shouldn't waste time making sure the font and colors and such are just right. Although choosing the right chart type will probably help you understand you own data as well. But when you are presenting to a stakeholder, all of this matters a lot.

I think that sometimes in the analytics world people treat data visualization like an afterthought. Like "Okay, I did a ton of great programming and analysis work, it took me weeks, I found some super cool things, and now I'm going to throw it into some charts really quick to make my presentation slides look pretty and official."

But the truth is that business stakeholders really at the core don't really care about (or often understand) the fancy programming you did or your analytic work unless it results in some kind of actionable insight for them and their business. So because of this, the difference between a well-made visualization and a poorly-made (or God-forbid a non-existent) visualization can be the difference between your stakeholder understanding what you are trying to say, realizing its amazing, and implementing your suggestions, or being completely lost or uninterested and choosing not to do anything because they don't really see the point.

So even if it takes you a good amount of time to get the visualzation just right, that time is worth it because it is what makes all of the analysis work that you did the past several weeks worth it.

# So Visualization is Important, How Do I Do It Well? 

### <span style="color:green">Do</span>

* Make visualizations!

This may sound like 'duh of course', but I have been in work situations where someone is tasked with getting a certain insight out of a certain dataset to share with the stakeholders, and they come back with a dataframe of numbers as the answer to show them. Stakeholders don’t want to see lines of numbers. Don’t do that. 

* Make visualizations that are clean and crisp (color is important here)

* Stick to the most important information

* Ask yourself why you are making this chart, who the audience is, and what the story is that you are trying to tell

* Experiment with different chart types to help you decide which one tells the story best

### <span style="color:red">Don't</span>

* Convey your insights as just numbers and words 

e.g. "55% of the customers like chocolate, 30% like cookies and cream, 10% like vanilla, and 5% like strawberry. Oh also the total number of customers surveyed was 74"

* Make visualization overly busy

* Show as much information as possible, everything that you found out about the data

* Assume whoever is looking at you data knows everything that you do

* Use the first chart type that comes to mind

# The McCandless Method

David McCandles is an author and creator of the site *informationisbeautiful.net* and the creator of some really beautiful published visualizations. He defines 4 elements that are necessary for a "good" visualization (see *informationisbeautiful.net/visualizations/what-makes-a-good-data-visualization/*) :
1. **Information** - The data! It should be accurate, consistent, and honest (no fudging numbers)
2. **Story** - The concept. Why is what you are trying to say relevant and meaningful? Why should the audience care? 
3. **Goal** - The 'Why' behing the picture. What is the purpose? 
4. **Visual Form** - The structure of your viz. Is it easy to interpret? Is it clear?

# Scatter Plot

Shows the relationships between different variables (usually 2 unless size or color of the marker is used as the 3rd variable)

```python
img = mpimg.imread('Scatterplot Example.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.data-to-viz.com/graph/scatter.html*

### <span style="color:green">Do</span>
* Show subgroups if present

### <span style="color:red">Don't</span>
* Don't overplot. Some scatter plots can just get too busy to be useful

# Line Chart

Used to track changes over periods of time. Can be used to compare changes over the same period of time for more than one group.

```python
img = mpimg.imread('Line Graph Example.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.data-to-viz.com/graph/line.html*

### <span style="color:green">Do</span>

* If you aren't starting you y-axis at 0, make sure to explicitely call it out so as not to be misleading

### <span style="color:red">Don't</span>

* Don't include too many lines on the same chart. Keep it readable.
* Don't use dual axis. This is often misleading. 
* Use grid lines unless it is actually helpful. They often add unnecessary busyness to the visual

# Column/Bar Chart

```python
img = mpimg.imread('Bar Chart Example.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.data-to-viz.com/graph/barplot.html*

```python
img = mpimg.imread('Stacked Bar Chart Example.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.data-to-viz.com/graph/barplot.html*

# Histogram

Used to see the distribution of one (or a few) variables

```python
img = mpimg.imread('Histogram Example.png')
fig = plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.data-to-viz.com/graph/histogram.html*

### <span style="color:green">Do</span>
* Try out different bin sizes

### <span style="color:red">Don't</span>
* Don't compare more than 3 or 4 variables on the same chart - gets too busy
* Don't use unequal bin widths, this skews the data

# Pie Chart

Used to show different segments that are a part of a whole. 

```python
img = mpimg.imread('Pie Chart Example.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Chart from https://www.coursera.org/learn/visualize-data/supplement/j9Wdl/the-wonderful-world-of-visualizations*

### <span style="color:green">Do</span>
* Avoid pie charts in general. Humans are bad at reading angles, and a better story can usually be told with bar plots.

### <span style="color:red">Don't</span>
* Use 3D features (really you should probably never use 3D features in plots)
* Have segments that don't sum to 100 or the 'whole population'

# Choosing a Chart Type

Some example questions to ask youself:
* What type of data do you have? (e.g numeric, categoric, time series, etc.)
* How many variables do you have? 
* Are you measuring changes over time? 
* Are there relationships between the data that need to be shown? 
* How many data points need to be shown? 
* Is the data ordered?

# Matplotlib

Yes - matplotlib can be a little confusing. If you use the internet for reference (which you totally should by the way) then you may have already seen by now that there are often several ways to do things. There are actually two main ways:
* Object-oriented
* Pyplot interfaces

Don't worry, you are not alone, this confused me forever. So let me see if I can make it a little clearer for you all. First here is some common jargon used by both methods:

Terms to remember:

**Figures** : The whole 'canvas' where you create your visualizations. A single figure can contain one or more Axes.

**Axes** : An area where points can be specified in terms of x-y coordinates (basically an area where a specific plot will be). Usually includes two or more Axis. Most objects are assigned to a certain axes. (Yes it's confusing that Axes and Axis are so similar, don't ask me why they did this)

```python
img = mpimg.imread('The Anatomy of a Figure.png')
fig = plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

*Picture from https://matplotlib.org/stable/tutorials/introductory/usage.html*

# Object-Oriented (Flexible and Customizable)

This is the approach where you explicitely create Figures and Axes (the 'objects'), and then call methods on them. This is the way the Matplotlib suggests for general use because it can be used to make more complicated plots, functions, and scripts. It is more flexible, and you have more control over every part of your visualization

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.5))
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

# PyPlot Interface (Quick and Simple)

Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.5))
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
```

Alot is similar, but some things are subtly different. This can make it hard to remember just how to do things if you switch back and forth from one method to another. So when in doubt, don't be afraid to Google it!

# Customizing Your Plot

```python
dataset1, dataset2, dataset3, dataset4 = np.random.randn(4, 100) # make 4 random data sets
```

```python
# Note: these are NOT examples of well-created plots, just showing how to customize things with matplotlib
fig, ax = plt.subplots(2, 1, figsize=(8, 3))
x = np.arange(len(dataset1))
ax[0].plot(x, np.cumsum(dataset1), color='purple', linewidth=2, linestyle='--', label='data1') # specify the color as a work, and the linestyle
ax[1].plot(x, np.cumsum(dataset2), marker='>', color='C0', markersize=3, linewidth=.5, label='data2') # specify the color as a colorcode, add markers
ax[1].set_xlabel('X Label', fontsize=14, color='red')
ax[1].set_ylabel('Y Label')
ax[0].set_title('A Really Cool Plot', fontsize=20)
ax[1].grid(True)
ax[1].annotate("Look here!", xy=(20,-15), xytext= (25, -9), arrowprops=dict(facecolor='black', shrink=0.05))
ax[0].legend() # add the legend, which includes the labels specified for the plots
plt.show()
```

# Seaborn

Seaborn is a library that was built on top of matplotlib for the purpose of making plots 'easier' to make and more visually appealing out-of-the-box. This means that it is pretty simple to use both seaborn and matplotlib together. Think of seaborn as giving you a simple and quick way to initially create the graph you want with some high level customizations available, and then you can use matplotlib to customize it further to get it exactly how you want. 

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("iris") # anything that starts with 'sns' is seaborn functionality
# Making one plot
sns.set_palette("vlag") # quick way to set the colors and 'make it pretty'
plt.subplot(211) # defines where we want out subplots to be - '2' subplot, in '1' column, and this is the first (1) plot
plt.title('Title using Matplotlib Function') # anything that starts with 'plt' is matplotlib functionality
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.plot()
# Making another plot
sns.set_palette("Accent") # quick way to set all colors to a certain palette
plt.subplot(212)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.plot()
plt.show()
```
